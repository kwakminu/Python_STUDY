**함수**

```python
def hello() :
  name = input("이름을 입력하세요! ")
  print("안녕!", name)

hello()
```

```python
def hello(name) :
  print("안녕!", name)

my_name = input("이름을 입력하세요! ")
hello(my_name)
```

```python
def happyBirthday(name):
  print("생일을 축하합니다!")
  print("사랑하는",name, "님의 생일을 축하합니다!")

happyBirthday("김숭실")
```

```python
num = 0

def change_num() :
  global num
  num = 100
  print(num)

change_num()
print(num)
```
