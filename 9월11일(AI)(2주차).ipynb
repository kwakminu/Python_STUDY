```python
while True :
  num = input("숫자 : ")
  if not num.isdigit()  :
    break

  num = int(num)
  if num > 0 :
   print("양수")
  elif num < 0 :
    print("음수")
  else :
    print("0")
```

숭실대 2023-2 2주차 연습문제

1번. 숫자가 입력 되는 동안 값을 입력 받고, 문자가 입력되면 '실행 종료'를 출력하고 총 합을 출력하시오.

```python
hap = 0
while True :
  num = input("값을 입력하세요 ")
  if not num.isdigit() :
    print("실행 종료")
    print("총합:", hap)
    break
  num = int(num)
  hap = hap + num
```
