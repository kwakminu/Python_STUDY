```python
import pandas as pd
data = pd.read_excel("/content/경찰청_카메라등이용촬영범죄 발생 피해자와의 관계 현황_20221231.xlsx")
data
```

```python
data2022 = data.drop(data.index[0:8], axis = 0)
data2022.reset_index(drop=True, inplace = True)
data2022.drop(columns = ['국가', '공무원', '구분'], axis = 1, inplace = True)
data2022
```

```python
data2022면식범 = data2022.drop(columns = ['비면식범_타인', '비면식범_기타'], axis=1)
data2022면식범
```

```python
data2022비면식범 = data2022.drop(columns = ['면식범_고용관계',	'면식범_직장동료',	'면식범_친구',	'면식범_애인',	'면식범_동거친족',	'면식범_기타친족',	'거래상대방',	'면식범_이웃',	'면식범_지인 등'], axis = 1)
data2022비면식범
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

면식범 = ['고용관계',	'직장동료',	'친구',	'애인',	'동',	'기', '거', '이','지인 등']
wedgeprops = {'width' : 0.6, 'linewidth' : 1, 'edgecolor' : 'black'}
arr = np.array(data2022면식범,dtype=int)
arr=arr.flatten()
plt.pie(arr, labels=면식범, autopct = "%d",wedgeprops = wedgeprops, shadow = True)
plt.title("피해자-가해자 관계(면식범)")
plt.show()
```

```python
import numpy as np

비면식범 = ['타인', '기타']
arr = np.array(data2022비면식범,dtype=int)
arr=arr.flatten()
plt.pie(arr, labels=비면식범, autopct = "%d",wedgeprops = wedgeprops, colors = ['cyan', 'white'], shadow = True)
plt.title("피해자-가해자 관계(비면식범)")
plt.show()
```

```python
이름 = ['면식범', '비면식범']
숫자 = [325, 1587]

plt.pie(숫자, labels = 이름, autopct="%d", shadow = True,wedgeprops = wedgeprops, colors=['white','pink'] )
plt.title("피해자-가해자 관계")
plt.show()
```
