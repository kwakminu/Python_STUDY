```python
#1주차 인강에 머신러닝이 뭐야 어떻게 돌아가는지? 개념은 쓸 수 있게 하라 : 데이터를 준비하고 그 데이터에 따라서 어떤 학습을 시켜서 그 학습시킨 결과를 가지고 새로운 데이터를 판단해주는 형태로 머신러닝을 할 수 있다.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
x2 = [1.4, 2.4, 3.4]
y = [30, 66, 23]
y2 = [45, 89, 8]
#x와 y의 개수가 같아야함!
#pie는 다른 그래프들과는 다르게 데이터 하나만 넣으면 된다! (원 그래프는 x축 y축이 없으니까)

plt.bar(x,y, alpha = 0.5, width = 0.3)#알파는 투명도 #width는 너비
plt.bar(x2, y2, width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,4,1)
x2 = x + 0.5
y = [30, 66, 23]
y2 = [45, 89, 8]


plt.bar(x,y, alpha = 0.5, width = 0.3)
plt.bar(x2, y2, width=0.2)
plt.show()
```
