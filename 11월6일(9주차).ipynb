```python
import pandas as pd
df = pd.read_excel("/content/전국도시공원정보표준데이터-20231106 (1).xls", header = [1]) #0번째 줄이 아니라 1번째 줄이 헤더가 됨
df = df.dropna(subset=["소재지지번주소", "위도", "경도"]) #소재지지번주소랑 위도, 경도가 빈칸이면 삭제해주는 함수 그렇구나~
df = df.query("소재지지번주소.str.contains('제주')",engine = 'python')

m = folium.Map(location=[37.4969, 126.9544], zoom_start=17)
for n,row in df.iterrows() : #따라 쓰면 됨 이해하려고 하지 않기
  folium.Marker([row['위도'], row['경도']], popup=row['공원명']).add_to(m) #위도 경도엔 대괄호를 다시 묶어주기

m
```
